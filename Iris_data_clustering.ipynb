```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset=pd.read_csv('iris_csv.csv')
X=dataset.iloc[:,0:4].values
```

**Elbow method**

```
from sklearn.cluster import KMeans
WCSS=[]
for i in range (1,10):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(X)
  WCSS.append(kmeans.inertia_)
plt.plot(range(1,10),WCSS)
plt.title('THE ELBOW METHOD')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()
```

**Silhouette Coefficient**

```
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
silhouett_coeff=[]
for i in range (2,10):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42).fit(X)
  y=kmeans.predict(X)
  silhouette = silhouette_score(X,y)
  silhouett_coeff.append(silhouette)
    
plt.plot(range(2,10),silhouett_coeff)
plt.title('Silhouett')
plt.xlabel('No. of clusters')
plt.ylabel('Silhouett scores')
plt.show()
```

**Davies Bouldin score**

```
from sklearn.metrics import davies_bouldin_score
db_score=[]
for i in range (2,10):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42).fit(X)
  y=kmeans.predict(X)
  db = davies_bouldin_score(X,y)
  db_score.append(db)
    
plt.plot(range(2,10),db_score)
plt.title('Davies_Bouldin score')
plt.xlabel('No. of clusters')
plt.ylabel('db scores')
plt.show()
```

**Training the K_Means model on the dataset**

```
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

**Visualisation**

```
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=80,c='red',label='cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=80,c='blue',label='cluster 2')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=80,c='green',label='cluster 3')
plt.xlabel('sepal-length')
plt.ylabel('sepal-width')
plt.title('clusters')
```
